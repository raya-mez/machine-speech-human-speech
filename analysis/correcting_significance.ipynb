```r
# import p-values from files
p_values_explanations <- read.csv("p_values_explanations.csv")
p_values_features <- read.csv("p_values_features.csv")

p_values <- rbind(p_values_explanations, p_values_features)

# adjust p-values for multiple comparisons using holm method
p_values$p_value_adj <- p.adjust(p_values$p_value, method = "holm")
# add significance column
p_values$significant <- p_values$p_value_adj < 0.05
# add level of significance column
p_values$significance_level <- cut(
    p_values$p_value_adj,
    breaks = c(-Inf, 0.001, 0.01, 0.05, 1),
    labels = c("***", "**", "*", "ns")
)

# print p-values with decimal notation (no scientific notation)
p_values_print <- p_values
p_values_print$p_value <- formatC(p_values_print$p_value, format = "f", digits = 4)
p_values_print$p_value_adj <- formatC(p_values_print$p_value_adj, format = "f", digits = 4)
print(p_values_print)
```
